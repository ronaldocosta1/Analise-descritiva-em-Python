####  <font color=#64ad49 >1.1 - Importando a base de dados</font>

```python
import numpy as np
import pandas as pd
```

```python
dados = pd.read_csv('State_of_Data_2021.csv', sep = ",")
# pd.set_option('display.max_rows',10)
# pd.set_option('display.max_columns', 20)
dados.head()
```

<font color=#b82a2a >É possivel observar que os dados brutos contém muitas variáveis e seus nomes estão desformatados</font>

####  <font color=#64ad49 >1.2 - Selecionando as colunas de interesse e mudando os nomes das variáveis</font>

```python
dados_news = dados.iloc[:,1:15] #Selecionando colunas
dados_news.head()
```

```python
dados_news.columns = ["idade", "faixa_idade", "genero","estado_onde_mora","uf onde mora",
                      "regiao_onde_mora","regiao_de_origem","mudou_de_estado","nivel_de_ensino",
                      "area_de_formação","qual_sua_situação_atual_de_trabalho?","setor",
                      "numero_de_Funcionarios","gestor"]  # Novos nomes

dados_news.head()
```

####  <font color=#64ad49 >1.3 - Informações gerais da base de dados</font>

```python
dados_news.info()

# Podemos observar: 
#   o número de variáveis;
#   o nome das variáveis;
#   quantos registros são 'não nulos';
#   seu tipo. 
```

#### <font color=#64ad49 >1.4 - Fazendo uma tabela mais 'bonita' e que  contém a descrição das variáveis</font>

```python
dic_var = ["Idade","Faixa de idade em que se encontra atualmente","Gênero","Estado onde mora",
           "UF onde mora","Região onde mora","Região de origem","Mudou de estado?","Nivel de formação",
           "Área em que se formou","Situação de trabalho","Setor em que trabalha",
           "Número de funcionarios do local em que trabalha","É gestor?"]
```

```python
tipos_de_dados = pd.DataFrame(dados_news.dtypes, columns = ["Tipos de Dados"])   # Para o gestor 
tipos_de_dados.columns.name = ""
tipos_de_dados["Descrição"] = np.array(dic_var)
tipos_de_dados = tipos_de_dados.rename_axis('Variáveis').reset_index()
tipos_de_dados
```

```python
linhas = dados_news.shape[0]
colunas = dados_news.shape[1]
print(f"O banco de dados contém {colunas} variáveis e {linhas} registros.")
```
